```python
def Heis_local_Ham_open_tree_dmrg(N_x,N_y,data_type="float64"):
 Z = qu.pauli('Z',dtype=data_type) * (0.5)
 X = qu.pauli('X',dtype=data_type) * (0.5)
 I = qu.pauli('X',dtype=data_type) * (0.5)
 Y_p = qu.pauli('Y') * (0.5)

 Y=np.array([[0, -1],[1,0]]) * (0.5)


 
 ZZ = Z &Z
 YY = Y_p & Y_p
 XX = X & X
 H_ham=0.25*( ZZ+XX+YY )
 e=qu.eigensystem(H_ham, True,k=-1, sort=True, return_vecs=False)
 Landa=max(e, key=abs)
 I=I*(abs(Landa)**(0.5))
# I=I*(3.)


 X=X.astype(data_type)
 Z=Z.astype(data_type)
 Y=Y.astype(data_type)
 I=I.astype(data_type)

 print (Landa, I,X,Y,Z)

    
 list_sites=[]
 list_inter=[]
 ##print (qu.ham_heis(2))



 for i in range(N_x-1): 
  for j in range(N_y): 
   list_sites.append( [ ( i,j), ((i+1),j) ]   )
   

 for i in range(N_x): 
  for j in range(N_y-1): 
   list_sites.append( [ ( i,j), (i, (j+1)) ]   )


 list_inter=[X,Z,Y,I]
 return list_sites, list_inter
```

```python
list_sites, list_inter=quf.Heis_local_Ham_open_tree(N_x,N_y,data_type=data_type)
opt="auto-hq"
print (quf.energy_f_tree(tn_U, list_sites, list_inter))
tnopt_mera=quf.auto_diff_tree(tn_U, list_sites,list_inter , opt, optimizer_c='L-BFGS-B')
tnopt_mera.optimizer = 'L-BFGS-B' 
tn_U = tnopt_mera.optimize(n=500 ,ftol= 2.220e-10, maxfun= 10e+9, gtol= 1e-12, eps= 1.49016e-08, maxls=400, iprint = 0, disp=False)
```

```python
list_sites, list_inter_tree=Heis_local_Ham_open_tree_dmrg(N_x,N_y,data_type="float32")
for count,tag_I   in enumerate(list_tags_I):
    print (count,tag_I,quf.energy_f_tree(tn_U, list_sites, list_inter), (tn_U&tn_U.H)^all )
    list_env=[]
    for iter in range(len(list_sites)):
        where=list_sites[iter]
        coo1,coo2=where
        tags = [tn_U.site_tag(*coo) for coo in where]
        tn_ij = tn_U.select(tags, which='any')
        if tag_I in tn_ij.tags:
              for count,ele in enumerate(list_inter_tree):
                      tn_ij_X=tn_ij.gate(ele, (coo1,) )
                      tn_ij_XX=tn_ij_X.gate(ele, (coo2,))
                      tn_ij_h=tn_ij.H
                      tn_ij_h[tag_I].modify(tags="GGGGGG")
                      tn_ij_overlap = ( tn_ij_h |  tn_ij_XX)
                      t_temp, = tn_ij_overlap.select_tensors(tag_I, which='any')
                      L_inds=list(t_temp.left_inds)
                      R_inds = [ix for ix in t_temp.inds if ix not in L_inds]
                      LR_inds = L_inds + R_inds
                      t_temp.transpose_(*LR_inds)
                      t_temp.modify(inds=LR_inds,left_inds=L_inds)
                      tn_env = tn_ij_overlap.select(tag_I, which='!any')
                      t_env = tn_ij_overlap.contract(all,optimize=opt, output_inds=t_temp.inds)
                      #print (count)

                      if count <2 :
                        list_env.append(t_env)
                      else:
                        list_env.append(t_env*(-1.0))
                            
            #else:
               #print ("No")
    Y_env=sum ( list_env[ix].data  for ix in range(len(list_env))   )
    t_env=list_env[0]*1.0
    t=tn_U[tag_I]
    L_inds=list(t.left_inds)
    R_inds = [ix for ix in t.inds if ix not in L_inds]
    LR_inds = L_inds + R_inds
    t.transpose_(*LR_inds)
    t_env.modify(data=Y_env)
    t_env.modify(inds=t.inds)
    t.modify(inds=LR_inds,left_inds=L_inds)
    
    if len( list(t.inds) )==3:
            L_inds=list(t.left_inds)
            R_inds = [ix for ix in t.inds if ix not in L_inds]
            LR_inds = L_inds + R_inds
            t.transpose_(*LR_inds)
            t1,t2,t3=t.inds
            (u, s, v)=t_env.split(left_inds=[t1,t2], method='svd',get='tensors', absorb=None, cutoff=-1e-17, right_inds=[t3]) 
            U_update=( u @ v) *(-1.0) 
            t.modify(data=U_update.data,inds=LR_inds,left_inds=L_inds)

    elif len( list(t.inds) )==2:
            L_inds=list(t.left_inds)
            R_inds = [ix for ix in t.inds if ix not in L_inds]
            LR_inds = L_inds + R_inds
            t.transpose_(*LR_inds)
            t1,t2=t.inds
            (u, s, v)=t_env.split(left_inds=[t1,t2], method='svd',get='tensors', absorb=None, cutoff=-1e-17, right_inds=[]) 
            U_update=( u @ v) *(-1.0) 
            t.modify(data=U_update.data,inds=LR_inds,left_inds=L_inds)

    
    
    
    
```
