# find 'compressed' path

`Reusable{}` optimizers can be supplied to many different contractions - they only run once per contraction and then cache the result. You can supply an optional `directory` kwarg to cache these results to disk.

```python
copt = ctg.ReusableHyperCompressedOptimizer(
    #methods=[],
    chi=12,
    minimize='peak-compressed',
    max_repeats=400,
    progbar=True,
    directory="cashlab/"
)
#parallel='ray'
```

```python
tree = tn.contraction_tree(copt)
```

```python
#copt.last_opt.plot_trials()
```

```python
tree.plot_contractions()
```

```python
#tree.plot_span()
```

```python
tree.plot_rubberband(with_edge_labels=False,with_node_labels=False,figsize=(3, 3),    centrality=False,
    layout=None,
    node_size=2,
    node_color=(.5, .5, .5, 1.0),
    edge_alpha=1 / 3,
    edge_style='solid',
    hyperedge_style='dashed',
    draw_edge_labels=False,
)
```

The compressed peak contraction widths for this path at different $\chi$:

# run with no gauging as reference

```python
y_none = []

for chi in chis:
    y_none.append(
        tn.contract_compressed(
            copt, 
            max_bond=chi,
            cutoff=0.0,
        )
    )
#print (y_none)
#y_none=[i^all for i in y_none]
```

```python
print ( y_none )
y_err_none = [abs(1 - y / ex).item() for y in y_none]
y_err_none
```

# Run with basic compression ('canonization' / tree-gauge)

```python
y_basic = []

for chi in chis:
    y_basic.append(
        tn.contract_compressed(
            copt, 
            max_bond=chi,
            cutoff=0.0,
            canonize_distance=1,
            canonize_after_distance=1,
            compress_late=True,
            gauge_boundary_only=False,
        )
    )

y_basic1 = []

for chi in chis:
    y_basic1.append(
        tn.contract_compressed(
            copt, 
            max_bond=chi,
            cutoff=0.0,
            canonize_distance=2,
            canonize_after_distance=2,
            compress_late=True,
            gauge_boundary_only=False,
        )
    )
    

y_basic2 = []

for chi in chis:
    y_basic2.append(
        tn.contract_compressed(
            copt, 
            max_bond=chi,
            cutoff=0.0,
            canonize_distance=2,
            canonize_after_distance=0,
            compress_late=True,
            gauge_boundary_only=False,
        )
    )

y_basic3 = []

for chi in chis:
    y_basic3.append(
        tn.contract_compressed(
            copt, 
            max_bond=chi,
            cutoff=0.0,
            canonize_distance=4,
            canonize_after_distance=4,
            compress_late=True,
            gauge_boundary_only=False,
        )
    )

#y_basic=[i^all for i in y_basic]
```

```python
#print (y_basic[0])
y_err_basic = [abs(1 - (y / ex) ).item() for y in y_basic]
y_err_basic1 = [abs(1 - (y / ex) ).item() for y in y_basic1]
y_err_basic2 = [abs(1 - (y / ex) ).item() for y in y_basic2]
y_err_basic3 = [abs(1 - (y / ex) ).item() for y in y_basic3]

y_err_basic
y_err_basic1
y_err_basic2
y_err_basic3
```

# Run with 'full-bond' compression 

Reusing compressed contraction path finder for inner approx env contraction

```python
y_full = []
chis = [16]
for chi in chis:
    y_full.append(
        tn.contract_compressed(
            copt, 
            max_bond=chi, 
            compress_opts=dict(
                mode='full-bond',
                # how we form the bond environment
                env_method='contract_compressed',
                # extra options to that function, which here itself is a compressed_contract
                contract_compressed_opts=dict(
                    optimize=copt,
                    canonize_distance=2,
                    canonize_after_distance=2,
                    compress_late=True
                ),
            ),
        )
    )
```

```python
y_err_full = [abs(1 - y / ex).item() for y in y_full ]
y_err_full
```

# plot results

```python
import matplotlib.pyplot as plt
plt.plot(chis, y_err_none, marker='.', label='none')
plt.plot(chis, y_err_basic, marker='.', label='dis=1,1')
plt.plot(chis, y_err_basic1, marker='.', label='dis=2,2')
plt.plot(chis, y_err_basic3, marker='.', label='dis=4,4')
plt.plot(chis, y_err_basic2, marker='.', label='dis=2,0')
plt.plot(chis, y_err_full, marker='.', label='full-bond, dis=2,2')
plt.yscale('log')
plt.legend()
plt.show()
```
